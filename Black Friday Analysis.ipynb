FUNCTIONS

```python
#Tagging records as 'HIGH FOCUSSED' transaction where Purchase amount > 5000 in a New Column - "Priority"
```

```python
#by declaring a function > def()


def target(x):
    if x > 5000:
        return "High Focussed"
    else:
        return "General"
        
```

```python
df["Priority"] = df.Purchase_Amount.apply(target)
```

```python
df.head()
```

```python
#Lamda Function


df["Priority"] = df.Purchase_Amount.apply(lambda x: "High Focussed" if x>5000 else "General")
```
